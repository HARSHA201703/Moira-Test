```python
df3["Total_Sales"]=df3['QuantityOrdered']*df3['ItemPrice']
```

```python
df3=df3.drop_duplicates(subset='OrderId')
```

```python
import json
df3['PromotionDiscount'] = df3['PromotionDiscount'].apply(lambda x: json.loads(x) if isinstance(x, str) else x)
df3['Amount'] = df3['PromotionDiscount'].apply(lambda x: x['Amount'])
df3['Amount'] = pd.to_numeric(df3['Amount'])
```

```python
df3['Net_Sales']=df3['Total_Sales']-df3['Amount']
```

```python
df3=df3[df3['Net_Sales']>0 ]
```
